## Backstory

In the mystical realm of Aethoria, an ancient practice is still followed with much reverence and anticipation - the Monster Trading Card Duel. This unique form of dueling was established by the realm's ancient settlers as a non-violent alternative to the gruesome wars of old. The settlers tamed the realm's many mythical beasts, transforming them into magical cards with each bearing the power and health of the creature it represents. Today, the citizens of Aethoria, young and old, still partake in this dueling tradition to test their strategic acumen and their luck, and to honor their ancestors.

## Game Rules

1. **Starting the Game:** Each player begins the duel with a hand of two cards drawn randomly from a shared pile of monster cards.

2. **Monster Cards:** Each card in the game represents a monster from the realm, with two attributes - Power and Health. The Power determines the monster's strength in combat, while the Health represents its endurance. However, in this simplified version of the game, only the Power attribute is used.

3. **Gameplay:** Players take turns to play a card from their hand. In each turn, a player selects one of their cards to battle against a card chosen by the opponent. The card with the higher Power wins the duel, and the losing card is removed from the game. If both cards have the same Power, it's considered a draw, and both cards are removed.

4. **Winning the Game:** The game continues until one player has no more cards left in their hand. The player with cards remaining is declared the winner. If both players exhaust their cards simultaneously, the game ends in a draw.

5. **Strategy:** Players must decide which card to play in each turn wisely, as the Power of the card not only determines the outcome of the duel but also the player's remaining options in subsequent turns.

This game emphasizes strategic thinking, luck, and the understanding of when to play your strongest cards or sacrifice the weaker ones. The true master of Monster Trading Card Duel can turn the tide of a match with clever tactics and bold plays. May the most cunning duelist win!

```python
import random

# Each monster card is represented by a dictionary with name, power and health.
# This is the initial collection of monster cards.
monster_cards = [
    {"name": "Dragon", "power": 10, "health": 20},
    {"name": "Goblin", "power": 5, "health": 10},
    {"name": "Ogre", "power": 8, "health": 15},
    {"name": "Unicorn", "power": 7, "health": 12},
]

# Each player starts with an empty list of cards.
player_cards = []
computer_cards = []

# The game starts with each player drawing 2 cards.
for _ in range(2):
    if len(monster_cards) > 0:
        card = random.choice(monster_cards)
        player_cards.append(card)
        monster_cards.remove(card)
    if len(monster_cards) > 0:
        card = random.choice(monster_cards)
        computer_cards.append(card)
        monster_cards.remove(card)

while True:
    if len(player_cards) == 0 and len(computer_cards) == 0:
        print("No more cards left to play. It's a draw!")
        break
    elif len(player_cards) == 0:
        print("You have no cards left. The computer wins!")
        break
    elif len(computer_cards) == 0:
        print("The computer has no cards left. You win!")
        break
    else:
        player_card = random.choice(player_cards)
        computer_card = random.choice(computer_cards)

        print(f"You play {player_card['name']} with power {player_card['power']} and health {player_card['health']}.")
        print(f"Computer plays {computer_card['name']} with power {computer_card['power']} and health {computer_card['health']}.")

        if player_card['power'] > computer_card['power']:
            print("Your card is stronger. You win this round!")
            computer_cards.remove(computer_card)
        elif player_card['power'] < computer_card['power']:
            print("The computer's card is stronger. The computer wins this round!")
            player_cards.remove(player_card)
        else:
            print("It's a draw. Both cards are removed.")
            player_cards.remove(player_card)
            computer_cards.remove(computer_card)
```
