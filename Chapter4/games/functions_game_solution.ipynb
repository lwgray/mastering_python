```python
import random

# Secret word
# this code randomly choses a secret word from the list
secret_word = random.choice(["python", "maria", "larry", "anna", "max"])

# Use `print` functions to Welcome the player and explain the game
print("Welcome to the guessing game!")
print("You have to guess the secret word. I will tell you how many letters it has.")

# Use `print` function to Tell the player how many letters the secret word has.
# Use `len` function to determine the number of letters in the secret word
print("The secret word has", len(secret_word), "letters.")

# Use `input` function to get the Player's guess
guess = input("Please enter your guess: ")

# Game loop
while guess != secret_word:
    # Use `print` function say something If the guess is incorrect
    print("That's not correct. Try again!")
    # Use `input` function to get another guess from the player
    guess = input("Please enter your guess: ")

# If the guess is correct
print("Congratulations! You've guessed the secret word.")
```
