# Capital Conundrum - Methods


```python
# Define the dictionary with countries and their capitals
capitals = {
    "USA": "Washington, D.C.",
    "France": "Paris",
    "UK": "London",
    "India": "New Delhi",
    "China": "Beijing",
    "Brazil": "Brasilia"
}

# Select a random country-capital pair
import random
country, correct_capital = random.choice(list(capitals.items()))


while True:
    print(f"What is the capital of {country}?")
    guess = input("Your guess: ")

    # Compare the guess to the correct capital (case-insensitive)
    if guess.lower() == correct_capital.lower():
        print("That's correct! Well done.\n")
        break
    else:
        print("That's not correct. Try again.\n")
```
