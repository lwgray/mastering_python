## Fill in the blanks
(1): Create an empty dictionary named `word_collection`

(2): Create an empty list named `recent words`

(3): What function should we use to get user input in Python?

(4): This method is used to add an item to the end of a list.

(5): This list method is used to add multiple items (from another list) to the end of the list. You may have to search the internet for the answer.

(6): This dictionary method displays a list of a dictionary's keys.

(7): What is the technique called when we retrieve a subset of a list?

Answers:

(1): word_collection

(2): recent_words

(3): input

(4): append

(5): extend

(6): keys

(7): Slicing

```python
word_collection = {}
recent_words = []

while True:
    word = input("Enter a word: ")

    # Add to recent words
    recent_words.append(word)

    # Categorize and add to word collection
    first_letter = word[0].lower()
    if first_letter not in word_collection:
        word_collection[first_letter] = [word]
    else:
        word_collection[first_letter].extend([word])

    # View collection
    for letter,value in word_collection.items():
        print(f"{letter.upper()}: {', '.join(word_collection[letter])}")

    # View recent words
    print("Recent words:", ', '.join(recent_words[-3:]))
```
