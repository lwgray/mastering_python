# List Methods Demo

### Game: The Magic Bag

**Concept**: You have a magical bag that can hold items. You can put items into the bag, take them out, and see what's inside at any time.

**Actions**:
1. Add an item to the bag.
2. Remove the last item from the bag.
3. View items in the bag.
4. Quit the game.

- You'll interact with the game by entering a number corresponding to one of the actions.
- The game responds based on your choice, either updating the bag's contents or displaying them.

```python
bag = []
actions = [
    "Add an item to the bag.",
    "Remove the last item from the bag.",
    "View items in the bag.",
    "Quit the game."
]

while True:
    # Display available actions
    for i, action in enumerate(actions):
        print(f"{i + 1}. {action}")

    # Get user's choice
    choice = input("Choose an action (1-4): ")

    # Add item to the bag
    if choice == "1":
        item = input("Enter the item to add: ")
        bag.append(item)
        print(f"You added {item} to the bag.")
        print()

    # Remove last item from the bag
    elif choice == "2":
        if bag:
            removed_item = bag.pop()
            print(f"You removed {removed_item} from the bag.")
        else:
            print("You cannot remove an item because the bag is empty.")
        print()

    # View items in the bag
    elif choice == "3":
        if bag:
            items = ', '.join(bag)
            print(f"The bag contains: {items}")
        else:
            print("The bag is empty.")
        print()

    # Quit the game
    elif choice == "4":
        print("Thanks for playing!")
        break

    # Invalid choice
    else:
        print("Invalid choice. Please choose a number between 1 and 4.")
        print()
```
