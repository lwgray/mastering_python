```python
## **Solution**:
# Define the spaceships and their speeds
spaceships = {
    "Rocket": 5,
    "Shuttle": 7,
    "UFO": 9,
    "Starship": 8
}

def get_round_winner(player1_choice, player2_choice):
    """
    Determine the winner of a round based on spaceship speeds.

    Parameters
    ----------
    player1_choice : str
        The spaceship chosen by Player 1.
    player2_choice : str
        The spaceship chosen by Player 2.

    Returns
    -------
    str
        "Player 1" if Player 1's spaceship is faster, "Player 2" if Player 2's spaceship is faster,
        and "Draw" if both spaceships have the same speed.
    """
    if spaceships[player1_choice] > spaceships[player2_choice]:
        return "Player 1"
    elif spaceships[player1_choice] < spaceships[player2_choice]:
        return "Player 2"
    else:
        return "Draw"

def play_round():
    """
    Execute a round where both players choose a spaceship and the round winner is determined.

    Returns
    -------
    str
        Winner of the round ("Player 1", "Player 2", or "Draw").
    """
    player1_choice = input("Player 1, choose your spaceship: ")
    player2_choice = input("Player 2, choose your spaceship: ")

    return get_round_winner(player1_choice, player2_choice)

def play_game():
    """
    Execute the main game logic for 3 rounds and determine the overall game winner.

    Returns
    -------
    None
    """
    player1_score = 0
    player2_score = 0

    for _ in range(3):
        winner = play_round()

        if winner == "Player 1":
            player1_score += 1
        elif winner == "Player 2":
            player2_score += 1

    game_winner = get_game_winner(player1_score, player2_score)
    print(f"The winner of Space Race is: {game_winner}")

play_game()
```
