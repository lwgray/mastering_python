**Game Outline**: **"Number Duel"**

In "Number Duel", two players each choose a number between 1 and 10. The computer then picks a random operation (addition, subtraction, multiplication, or division). The operation is applied to the two numbers, and the player whose number is larger after the operation wins the round. The game goes on for 5 rounds, and the player with the most round wins at the end is declared the champion.

1. For the `apply_operation` function, based on the value of `operation`, apply the relevant mathematical operation to `num1` and `num2`.
2. For the `get_round_winner` function, write a docstring that explains its purpose.
3. Implement the logic in the `get_round_winner` function that determines and returns the winner based on `result1` and `result2`.

````python
import random

def choose_operation():
    """
    Choose a random operation among addition, subtraction, multiplication, and division.

    Returns
    -------
    str
        The chosen operation: 'add', 'subtract', 'multiply', or 'divide'.
    """
    operations = ['add', 'subtract', 'multiply', 'divide']
    return random.choice(operations)

def apply_operation(num1, num2, operation):
        """
    Apply the specified operation to the two numbers and return the results.

    Parameters
    ----------
    num1 : int or float
        The first number.
    num2 : int or float
        The second number.
    operation : str
        The operation to apply: 'add', 'subtract', 'multiply', or 'divide'.

    Returns
    -------
    tuple of floats : (float1, float2)
        The results of the operation applied to num1 and num2. The first element in the tuple
        should be the result of applying the operation to num1 and num2. The second element
        should be the result of applying the operation to num2 and num1.

        For example, if the operation is 'add', the first element should be (num1 + num2)
        and the second element should be (num2 + num1) (which will be the same in this case).
    """
    # Hint: Use if-elif-else statements to check what the operation is.
    #       Then perform that operation on num1 and num2.
    #       Don't forget to handle the division by zero case for the 'divide' operation!

    if operation == 'add':
        return (num1 + num2, num2 + num1)
    elif operation == 'subtract':
        return (num1 - num2, num2 - num1)
    elif operation == 'multiply':
        return (num1 * num2, num2 * num1)
    elif operation == 'divide':
        if num2 == 0 or num1 == 0:
          return ("Cannot divide by zero", "Cannot divide by zero")
        return (num1 / num2, num2 / num1)

def get_round_winner(result1, result2):
    """
    Determine the winner of a round based on the results of the applied operation.

    Parameters
    ----------
    result1 : float
        Result of the operation for Player 1's number.
    result2 : float
        Result of the operation for Player 2's number.

    Returns
    -------
    str
        Returns "Player 1" if Player 1's result is larger, "Player 2" if Player 2's result is larger, and "Draw" if both results are equal.
    """
    if result1 > result2:
        return "Player 1"
    elif result1 < result2:
        return "Player 2"
    else:
        return "Draw"


def play_round():
    """
    Execute a round where both players choose a number and the round winner is determined.

    Returns
    -------
    str
        Returns the winner of the round ("Player 1", "Player 2", or "Draw").
    """
    player1_num = float(input("Player 1, choose a number between 1 and 10: "))
    player2_num = float(input("Player 2, choose a number between 1 and 10: "))

    operation = choose_operation()
    print(f"Chosen operation for this round: {operation}")

    result1, result2 = apply_operation(player1_num, player2_num, operation)

    return get_round_winner(result1, result2)

def play_game():
    """
    Execute the main game logic for 5 rounds and determine the overall game winner.

    Returns
    -------
    None
    """
    player1_score = 0
    player2_score = 0

    for _ in range(5):
        winner = play_round()

        if winner == "Player 1":
            player1_score += 1
        elif winner == "Player 2":
            player2_score += 1

    if player1_score > player2_score:
        print("Player 1 is the champion!")
    elif player1_score < player2_score:
        print("Player 2 is the champion!")
    else:
        print("It's a draw!")

play_game()
````
