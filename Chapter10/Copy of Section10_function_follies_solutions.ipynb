### Exercise 1: Piggy Bank ðŸ·

```python
def piggy_bank(pennies, nickels, dimes, quarters):
  """
  Calculates the total amount in dollars from coins

  Parameters
  ----------
  pennies : int
    Number of pennies
  nickels : int
    Number of nickels
  dimes : int
    Number of dimes
  quarters : int
    Number of quarters

  Returns:
  float : Total amount in dollars
  """
  return pennies * 0.01 + nickels * 0.05 + dimes * 0.10 + quarters * 0.25
```

```python
piggy_bank(1, 1, 1, 1)
```

### Exercise 2: Movie Night ðŸ¿

```python
def movie_night(movies, genre):
  """
  Selects movies that match a specific genre

  Parameters
  ----------
  movies : list
    A list of movies (dictionaries with title and genre)
  genre : str
    The selected genre

  Returns:
  list : A list of movies that match the genre
  """
  matching_movies = []
  for movie in movies:
    if movie['genre'] == genre:
      matching_movies.append(movie)
  #[movie for movie in movies if movie['genre'] == genre]
  return matching_movies
```

```python
films = [{'title':'inception', 'genre':'Action'}, {'title':'Shrek', 'genre':'Animation'}]
movie_night(films, 'Action')
```

### Exercise 3: Weather Forecast â˜€ï¸â˜”


```python
def weather_forecast(temperature, condition):
  """
    Provides a clothing recommendation based on the weather.

    Parameters
    ----------
    temperature (int): The temperature in degrees Fahrenheit.
    condition (str): The weather condition (e.g., "sunny", "rainy").

    Returns:
    str: A recommendation on what to wear.
  """

  if condition == "rainy":
    return "Bring an umbrella!"
  elif temperature > 75:
    return "Wear shorts and sunglasses!"
  else:
    return "Bring a jacket"
```

```python
weather_forecast(80, "sunny")
```
