# Galactic Guardians Duel
##Game Overview:
In the vastness of the universe, a duel between cosmic entities ensues! The "Galactic Guardians Duel" is a card game where players use celestial cards, each representing a guardian of a galaxy with unique power levels and abilities.

Two players will each draw a card from their deck and play it. Each card has two attributes: Power and Ability. The card with the higher Power wins, but the Ability can modify the Power of its own or the opponent's card.

The rules for the abilities and powers in the "Galactic Guardians Duel" card game are as follows:

### Card Attributes:
1. **Power**: Every card has a Power value, which is a number between 1 to 10. This Power value determines the basic strength of the card.
2. **Ability**: Each card also has an Ability. There are three types of abilities: "Boost", "Drain", and "Shield".

### Abilities:

1. **Boost**:
    - Effect: Increases the card's Power by 2.
    - Example: If a card with a Power of 5 and the "Boost" ability is played, its effective Power becomes 7 for that round.

2. **Drain**:
    - Effect: Decreases the opponent's card Power by 2.
    - However, if the opponent's card has the "Shield" ability, then the "Drain" effect is nullified.
    - Example: If a card with a Power of 7 and the "Drain" ability is played against a card with a Power of 6, the opponent's card's Power decreases to 4 for that round, unless it has the "Shield" ability.

3. **Shield**:
    - Effect: Protects the card from the "Drain" ability of an opponent's card.
    - If the opponent plays a card with the "Drain" ability, the "Shield" prevents the Power reduction.
    - Example: If a card with a Power of 6 and the "Shield" ability is played against a card with the "Drain" ability, its Power remains 6 for that round.

### Gameplay:

1. Both players draw a card from their deck.
2. The abilities of the drawn cards are applied:
    - If a card has the "Boost" ability, its Power is increased by 2.
    - If a card has the "Drain" ability, the opponent's card Power is decreased by 2, unless the opponent's card has the "Shield" ability.
    - The "Shield" ability protects a card from the "Drain" effect of an opponent's card.
3. The card with the higher Power after applying the abilities wins the round.
4. If both cards have the same Power after applying the abilities, the round is a draw.
