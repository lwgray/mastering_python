# Mystery on Elm Street - Sample Solution

This is a sample solution for "Mystery on Elm Street," a text-based detective game.

```
def choose_location():
    """
    This function allows the player to choose a location to visit for gathering clues.

    Returns:
        str: The location chosen by the player.
    """
    options = ['Library', 'Cafe', 'Park']
    print("Choose a location to visit:")
    for i, option in enumerate(options):
        print(f"{i+1}. {option}")
    choice = int(input("Enter the number of your choice: "))
    return options[choice - 1]

def gather_clues(location):
    """
    This function gathers clues based on the location chosen by the player.

    Parameters:
        location (str): The location where the player wants to gather clues.

    Returns:
        list: A list of clues gathered.
    """
    clue_map = {
        'Library': ['Book on ancient rituals', 'Newspaper article about a missing person'],
        'Cafe': ['Receipt for two coffees', 'Note with a phone number'],
        'Park': ['Footprints', 'Broken watch']
    }
    return clue_map.get(location, [])

def interrogate_suspects():
    """
    This function allows the player to interrogate suspects to gather more information.

    Returns:
        str: Statement from the chosen suspect.
    """
    suspects = ['Alice', 'Bob', 'Charlie']
    statements = [
        'I was at the library all day.',
        'I haven\'t seen anything suspicious.',
        'I lost my watch in the park.'
    ]
    print("Choose a suspect to interrogate:")
    for i, suspect in enumerate(suspects):
        print(f"{i+1}. {suspect}")
    choice = int(input("Enter the number of your choice: "))
    return statements[choice - 1]

def make_accusation():
    """
    This function allows the player to make an accusation to solve the mystery.

    Returns:
        bool: True if the accusation is correct, False otherwise.
    """
    correct_answer = 'Alice'
    accusation = input("Who do you accuse? ")
    return accusation == correct_answer

def main():
    """
    The main function where the game loop resides.
    """
    while True:
        location = choose_location()
        clues = gather_clues(location)
        print(f"Clues gathered at {location}: {clues}")

        statement = interrogate_suspects()
        print(f"Suspect says: {statement}")

        accusation = make_accusation()
        if accusation:
            print("Congratulations, you solved the mystery!")
            break
        else:
            print("Wrong accusation. Keep investigating.")

main()
```
