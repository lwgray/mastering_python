In a Jupyter Notebook, write a code to demonstrate the use of the following data types:
- Integer
- Float
- String
- List

For each data type, write a code that does the following:
1. Create a variable with a relevant name and assign a value to it.
2. Print the data type of the variable.
3. Perform at least two operations on the variable (e.g., add, multiply, concatenate, etc.) and print the result of the operations




```python
#Integer:

# Create a variable
age = 27

# Print the data type of the variable
print(type(age))
# output: int

# Perform operations
age = age + 3 # age += 3
age = age * 2

# Print the result of the operations
print(age)
# Output: 60
```

```python
# Float

# Create a variable
height = 70.5 # in inches

print(type(height))

# Perform operations
height = height / 3.5 # division
height = height // 2 ## floor division

print(height)
```

```python
# String

# Create a variable
greeting = "hello "

print(type(greeting))

# Perform operations
greeting = greeting + "World" # concatenation
greeting = greeting * 2 # Repeat

print(greeting)
```

```python
# List
# 1. Creating a variable
my_list = [1, 2, 3]

# 2. Printing the data type
print(type(my_list))

# 3. Performing mathematical operations
# Concatenation
result1 = my_list + [4, 5]
print(result1)

# Repetition
result2 = my_list * 2
print(result2)
```
