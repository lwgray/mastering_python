# The Great Operator Safari:
Welcome to "The Great Operator Safari: A Quest for Relational Wisdom"! In this thrilling journey, you'll venture into the wild terrains of comparison operators, list evaluations, and object identity checks. This isn't your average stroll through Python; it's a quest to master the art of comparing values and understanding the nature of objects in Python!



## Exercises

1. **Basic Comparisons**:
   - Compare if `7` is greater than `5` and print the result.
   - Check if the string "hello" is equal to "Hello" and print the result.
   - Determine if `9` is not equal to `10` and print the result.



2. **List Comparisons**:
   - Given the list `fruits = ["apple", "banana", "cherry"]`, check if "mango" is in the list and print the result.
   - Using the same list, check if "banana" is not in the list and print the result.



3. **Object Identity**:
   - Create two lists `list_a = [1, 2, 3]` and `list_b = list_a`. Check if both lists point to the same memory location and print the result.
   - Now, create another list `list_c = [1, 2, 3]`. Check if `list_a` and `list_c` are the same object and print the result.
